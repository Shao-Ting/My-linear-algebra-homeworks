```python
%matplotlib inline

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

## 1.設一個2X2的矩陣A

```python
A = np.array([[5,4],[8,7]])
```

## 2.畫一個圖形

```python
π = np.pi
θ = np.linspace(0, 2*π, 600)
```

```python
x = np.cos(θ) + θ
y = np.sin(θ) + θ
```

#### 看看長怎樣

```python
plt.plot(x,y)
```

## 3.進行轉換

```python
X = np.array([x,y])
egg = A@X
```

```python
Xi = egg[0]
Yi = egg[1]
```

#### 看看轉換後的圖形

```python
ax = plt.gca()
plt.plot(Xi, Yi)
ax.set_aspect('equal')
```

## 4.比較看看轉換前後的圖形

```python
plt.subplot(121)
ax = plt.gca()
plt.plot(x,y)
ax.set_aspect('equal')
plt.xlim(0,15)
plt.ylim(0,15)

plt.subplot(122)
ax = plt.gca()
plt.plot(Xi,Yi)
ax.set_aspect('equal')
plt.xlim(0,15)
plt.ylim(0,15)
```
